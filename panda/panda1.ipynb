```python
import pandas as pd
import random

# List of sample city names
cities = ['New York', 'Los Angeles', 'Chicago', 'Houston', 'Phoenix', 
          'Philadelphia', 'San Antonio', 'San Diego', 'Dallas', 'San Jose']

# List of sample names
first_names = ['John', 'Jane', 'Alex', 'Emily', 'Michael', 'Sarah', 
               'David', 'Laura', 'Chris', 'Sophia']
last_names = ['Smith', 'Johnson', 'Williams', 'Brown', 'Jones', 
              'Garcia', 'Miller', 'Davis', 'Rodriguez', 'Martinez']

# Generate sample data
data = {
    'emp_id': [i for i in range(1, 101)],
    'name': [f"{random.choice(first_names)} {random.choice(last_names)}" for _ in range(100)],
    'city': [random.choice(cities) for _ in range(100)],
    'salary': [random.randint(50000, 150000) for _ in range(100)]
}


# Create DataFrame
df = pd.DataFrame(data)

# Display the first 10 rows of the DataFrame
df
```

```python
df.groupby('city')['salary'].mean()
```
