#### mfc: Marker Face Color. This sets the color of the interior of the marker. 'w' stands for white.
#### mec: Marker Edge Color. This sets the color of the edge of the marker. 'k' stands for black.
#### ms: Marker Size. This sets the size of the marker. 15 is the size of the marker.
#### 

```python
# LST = [1,2,3,4,5,6]
LST = [10,20,30,10,40,30]

plt.plot(LST,color="red", marker = 'o', mfc='w', mec='k', ms=10, label="Age", ls= "--")
plt.title('Line Plot', color="k", fontsize=25)
plt.xlabel('ID',fontsize=25)
plt.ylabel('Age',fontsize=25)
plt.legend()
plt.grid()
plt.show()
```

# Line Plot

```python
import matplotlib.pyplot as plt
import numpy as np

x = [1, 2, 3, 4, 5]

y1 = [50, 20, 60, 40, 10]
y2 = [20, 200, 40, 20, 50]
y3 = [10, 30, 40, 100, 60]
y4 = [30, 40, 10, 50, 80]

plt.figure(figsize=(12, 7))
plt.plot(x, y1, color='b', label="Royal Enfield", marker="o", lw=1)
plt.plot(x, y2, color='m', label="Himalyan", marker="o",lw=1)
plt.plot(x, y3, color='y', label="BMW GS-1200", marker="o",lw=1)
plt.plot(x, y4, color='r', label="Harley", marker="o",lw=1)
plt.title('Bike Detail using Line Plot', color="k", fontsize=25)
plt.xlabel('Days', fontsize=25)
plt.ylabel('Km', fontsize=25)
plt.xticks(np.arange(1, 6, step=1))
plt.yticks(np.arange(10, 201, step=20))
plt.legend()
plt.grid()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(len([1, 2, 3, 4, 5]))  # Days
y1 = [50, 20, 60, 40, 10]
y2 = [20, 200, 40, 20, 50]
y3 = [10, 30, 40, 100, 60]
y4 = [30, 40, 10, 50, 80]

# Define the width of each bar
bar_width = 0.2

plt.figure(figsize=(12, 7))

# Create bars
plt.bar(x, y1, width=bar_width, color='b', label="Royal Enfield", edgecolor='grey')
plt.bar(x + bar_width, y2, width=bar_width, color='m', label="Himalyan", edgecolor='grey')
plt.bar(x + 2 * bar_width, y3, width=bar_width, color='y', label="BMW GS-1200", edgecolor='grey')
plt.bar(x + 3 * bar_width, y4, width=bar_width, color='r', label="Harley", edgecolor='grey')

# Adding labels and title
plt.title('Bike Detail using Bar Plot', color="k", fontsize=25)
plt.xlabel('Days', fontsize=25)
plt.ylabel('Km', fontsize=25)

# Customizing the ticks
plt.xticks(x + 1.5 * bar_width, [1, 2, 3, 4, 5])
plt.yticks(np.arange(10, 201, step=20))

# Adding legend and grid
plt.legend()
plt.grid(axis='y')

# Show plot
plt.show()
```

```python

x = [1, 2, 3, 4, 5]
y1 = [50, 20, 60, 40, 10]
y2 = [20, 200, 40, 20, 50]
y3 = [10, 30, 40, 100, 60]
y4 = [30, 40, 10, 50, 80]

plt.figure(figsize=(12, 7))
plt.plot(x, y1, color='b', linestyle='--', linewidth=2, label="Royal Enfield", marker="o")
plt.plot(x, y2, color='m', linestyle='-.', linewidth=2, label="Himalyan", marker="o")
plt.plot(x, y3, color='y', linestyle=':', linewidth=2, label="BMW GS-1200", marker="o")
plt.plot(x, y4, color='r', linestyle='-', linewidth=2, label="Harley", marker="o")

# Adding annotations
for i in range(len(x)):
    plt.text(x[i], y1[i], f'{y1[i]}', ha='center', va='bottom')
    plt.text(x[i], y2[i], f'{y2[i]}', ha='center', va='bottom')
    plt.text(x[i], y3[i], f'{y3[i]}', ha='center', va='bottom')
    plt.text(x[i], y4[i], f'{y4[i]}', ha='center', va='bottom')

# Shaded region
plt.fill_between(x, y1, y2, color='gray', alpha=0.2, label="Range between Y1 and Y2")

plt.title('Bike Detail using Line Plot', color="k", fontsize=25)
plt.xlabel('Days', fontsize=25)
plt.ylabel('Km', fontsize=25)
plt.xticks(np.arange(1, 6, step=1))
plt.yticks(np.arange(10, 201, step=20))
plt.legend()
plt.grid()
plt.show()
```
