```python

x = [1, 2, 3, 4, 5]

y1 = [50, 20, 60, 40, 10]
y2 = [20, 200, 40, 20, 50]
y3 = [10, 30, 40, 100, 60]
y4 = [30, 40, 10, 50, 80]

plt.figure(figsize=(10,4))

plt.plot(x,y1, color="b", label = "Royal Enfield", marker="o",lw=1)
plt.plot(x,y2, color="r", label = "Himalyan", marker="o",lw=1)
plt.plot(x,y3, color="k", label = "Harley", marker="o",lw=1)
plt.plot(x,y4, color="g", label = "BMW GS-1200", marker="o",lw=1)

plt.title("Bike Detail using LINE PLOT")
plt.xlabel("Number of Days")
plt.ylabel("Distance Covered in KM")
plt.xticks(np.arange(1,6, step=1))
plt.yticks(np.arange(0,201, step=20))


plt.legend()
plt.grid()
plt.show()
```

```python

x = [1, 2, 3, 4, 5]
y1 = [50, 20, 60, 40, 10]
y2 = [20, 200, 40, 20, 50]
y3 = [10, 30, 40, 100, 60]
y4 = [30, 40, 10, 50, 80]

plt.figure(figsize=(12, 7))
plt.plot(x, y1, color='b', linestyle='--', linewidth=2, label="Royal Enfield", marker="o")
plt.plot(x, y2, color='m', linestyle='-.', linewidth=2, label="Himalyan", marker="o")
plt.plot(x, y3, color='y', linestyle=':', linewidth=2, label="BMW GS-1200", marker="o")
plt.plot(x, y4, color='r', linestyle='-', linewidth=2, label="Harley", marker="o")

# Adding annotations
for i in range(len(x)):
    plt.text(x[i], y1[i], f'{y1[i]}', ha='center', va='bottom')
    plt.text(x[i], y2[i], f'{y2[i]}', ha='center', va='bottom')
    plt.text(x[i], y3[i], f'{y3[i]}', ha='center', va='bottom')
    plt.text(x[i], y4[i], f'{y4[i]}', ha='center', va='bottom')

# Shaded region
plt.fill_between(x, y1, y2, color='gray', alpha=0.4, label="Range between Y1 and Y2")

plt.title('Bike Detail using Line Plot', color="k", fontsize=25)
plt.xlabel('Days', fontsize=25)
plt.ylabel('Km', fontsize=25)
plt.xticks(np.arange(1, 6, step=1))
plt.yticks(np.arange(10, 201, step=20))
plt.legend()
plt.grid()
plt.show()
```
